```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('/content/Workbook_1.xlsx')
df2=pd.read_excel('/content/Workbook_2.xlsx')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['Name'].isin(df2['Name'])
```

```python
filtered_df1=df1.loc[df1['Name'].isin(df2['Name'])]
filtered_df1
```

Filter for everyone that needs to complete thegroup interview process


```python
filtered_df2=df1.loc[~(df1['Name'].isin(df2['Name']))]
filtered_df2
```

Candidates that meet certain conditions

```python
filtered_df3=df1.loc[(df1['Name'].isin(df2['Name'])) & (df1['Interview Score']>4)]
filtered_df3
```

Merge

```python
all_df=pd.merge(df1, df2, how='outer', on='Name' )
all_df
```

```python
filtered_df_4=all_df.loc[(all_df['YR Experience']<5) & (all_df['Group Interview Score']>4)]
filtered_df_4
```

```python
filtered_df_4.to_excel('final_candidtes.xlsx', index=False)
```
