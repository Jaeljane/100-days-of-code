```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/If statement.xlsx')
df.head()
```

```python
df['Average']=df.mean(axis=1)
```

```python
df['Pass/Fail']=np.where(df['Average']>40, 'Pass', 'Fail')
df
```

```python
conditions=[
(df['Average']>=80),
(df['Average']<70)&(df['Average']>=60),
(df['Average']<60)&(df['Average']>=50),
(df['Average']<50)&(df['Average']>=40),
(df['Average']<40)]

results=['A', 'B', 'C', 'D', 'E']
```

```python
df['Grade']=np.select(conditions, results)
df
```

```python
df['Pass or fails']=['Pass' if x> 40 else 'fail' for x in df['Average']]
df
```
